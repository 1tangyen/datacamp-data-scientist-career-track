## Broadcasting in arithmetic formulas

```python
temps_f = weather[['Min TemperatureF', 'Mean TemperatureF', 'Max TemperatureF']]

# Convert temps_f to celsius: temps_c
def f2c(f):
    return (f-32)*5/9
    
temps_c = temps_f.apply(f2c)
    
# Rename 'F' in column names with 'C': temps_c.columns
temps_c.columns = temps_c.columns.str.replace('F', 'c')

# print first 5 rows of temps_f, temps_c
print(temps_f.head())

print(temps_c.head())
```

```python
#Initialize an empyy list: medals
medals =[]
medal_types = ['bronze', 'silver', 'gold']

for medal in medal_types:
    # Create the file name: file_name
    file_name = [bronzetop5, silvertop5, goldtop5]
    # Create list of column names: columns
    columns = ['Country', medal]
    # Read file_name into a DataFrame: medal_df
    medal_df = pd.read_csv(file_name, header=0, index_col='Country', names=columns)
    # Append medal_df to medals
    medals.append(medal_df)

# Concatenate medals horizontally: medals_df
medals_df = pd.concat(medals, axis='columns')

# Print medals_df
print(medals_df)
```

# 4 Case Study - Summer Olympics

```python
# loading data

allmedalists1896_2008_url = 'https://raw.githubusercontent.com/iDataist/Merging-DataFrames-with-pandas/master/Summer%20Olympic%20medals/Summer%20Olympic%20medalists%201896%20to%202008%20-%20ALL%20MEDALISTS.tsv'

editions1896_2008_url = 'https://raw.githubusercontent.com/iDataist/Merging-DataFrames-with-pandas/master/Summer%20Olympic%20medals/Summer%20Olympic%20medalists%201896%20to%202008%20-%20EDITIONS.tsv'

IOCcountrycodes1896_2008_url = 'https://raw.githubusercontent.com/iDataist/Merging-DataFrames-with-pandas/master/Summer%20Olympic%20medals/Summer%20Olympic%20medalists%201896%20to%202008%20-%20IOC%20COUNTRY%20CODES.csv'

import pandas as pd
```

```python

editions = pd.read_csv(editions1896_2008_url, sep='\t')
editions = editions[['Edition', 'Grand Total', 'City', 'Country']]
editions.head()
```

```python
ioc_codes = pd.read_csv(IOCcountrycodes1896_2008_url)
ioc_codes = ioc_codes[['Country', 'NOC']]
ioc_codes.head()
```

```python
medals = pd.read_csv(allmedalists1896_2008_url, sep='\t', header=4)
medals = medals[['Athlete', 'NOC', 'Medal', 'Edition']]
medals.head()
```

```python
# construct the pivot_table: medal_counts
medal_counts = medals.pivot_table(index='Edition', columns='NOC', values='Athlete', aggfunc='count')

medal_counts.head()
```

```python
medal_counts.tail()
```

## computing fraction of medals per Olympic edition

```python
# set index of editions: totals
totals = editions.set_index('Edition')
totals.head()
```

```python
# reassign total['Grand Total']: totals
totals = totals['Grand Total']
totals.head()
```

```python
#Divide medal_counts by totals:fractions
fractions = medal_counts.divide(totals, axis='rows')

fractions.head()
```

```python
fractions.tail()
```

```python
# apply the expanding mean: mean_fractions
mean_fractions = fractions.expanding().mean()
mean_fractions.head()
```

```python
# compute the percentage change: fractions_change
fractions_change = mean_fractions.pct_change()*100
fractions_change.head()
```

```python
fractions_change = fractions_change.reset_index()
fractions_change.head()
```

```python
fractions_change.tail()
```

## Reshaping and plotting

```python
# building hosts DataFrame
#left join editions and ioc_codes: hosts
hosts = pd.merge(editions, ioc_codes, how='left')
hosts.head()
```

```python
# extract relevant columns and set index: hosts
hosts = hosts[['Edition', 'NOC']].set_index('Edition')
hosts.head()
```

```python
# fixing missing 'NOC' values of hosts
hosts.loc[hosts.NOC.isnull()]
```

```python
hosts.loc[1972, 'NOC'] = 'FRG'
hosts.loc[1980, 'NOC'] = 'URS'
hosts.loc[1988, 'NOC'] = 'KOR'
```

```python
hosts.reset_index(inplace=True)
hosts.head()
```

```python
# merge reshaped and hosts: merged
reshaped = pd.melt(fractions_change, id_vars='Edition', value_name='Change')

merged = pd.merge(reshaped, hosts, how='inner')

merged.head()
```

```python
# set index of merged and sort it: influence
influence = merged.set_index('Edition').sort_index()

influence.head()
```

## Plotting influence of host country

```python
import matplotlib.pyplot as plt
# Extract influence['Chnage']: change
change = influence['Change']

ax = change.plot(kind='bar')

ax.set_ylabel('% Change of Host Country Medal Count')
ax.set_title('Is there a Host Country Advantage')
ax.set_xticklabels(editions['City'])

plt.show()
```
