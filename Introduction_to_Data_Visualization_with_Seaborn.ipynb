# **Introduction to Seaborn**
# Introduction
## Making a scatter plot with lists

```
# Import Pandas, Matplotlib and Seaborn
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

plt.rcParams['figure.figsize'] = [10, 5]
```

```
# Loading data
df = pd.read_csv(countries_url)

df.head()
```

```
gdp = df['GDP ($ per capita)'].tolist()
phones = df['Phones (per 1000)'].tolist()
percent_literate = df['Literacy (%)']
region = df['Region'].tolist()
```

```
# Change this scatter plot to have percent literate on the y-axis
sns.scatterplot(x=gdp, y=phones)

# Show plot
plt.show()
```

```
# Change this scatter plot to have percent literate on the y-axis
sns.scatterplot(x=gdp, y=percent_literate)

# Show plot
plt.show()
```

## Making a count plot with a list

```
# Create count plot with region on the y=axis
sns.countplot(y=region)

plt.show()
```

# Adding a third variable with hue
## Hue and scatter plots

```
student_data = pd.read_csv(students_url)
student_data.head()
```

```
# Create a scatter plot of absences vs. final grades
sns.scatterplot(x='absences', y='G3', data=student_data, hue='location')

plt.show()
```

```
# Change the legend order in the scatter plot
sns.scatterplot(x='absences', y='G3',
                data=student_data, 
                hue='location', hue_order=['Rural', 'Urban'])

plt.show()
```

## Hue and count plots

```
# Create a dictionary mapping subgroup values to colors
palette_colors = {'Rural':'green', 'Urban':'blue'}

# Create a count plot of school with location subgroups
sns.countplot(x='school',
              data=student_data, 
              hue='location', palette=palette_colors)

plt.show()
```

# **Visualizing Two Quantitative Variables**
# Introduction to relational plots and subplots
## Creating subplots with col and row

```
# Change to use relplot() instread of scatterplot()
sns.relplot(x='absences', y='G3',
            data=student_data,
            kind='scatter')

plt.show()
```

```
# Change to make subplots based on study time
sns.relplot(x='absences', y='G3',
            data=student_data,
            kind='scatter',
            col='study_time')

plt.show()
```

```
# Change this scatter plot to arrange the plots in rows instead of columns
sns.relplot(x='absences', y='G3',
            data=student_data,
            kind='scatter',
            row='study_time')

plt.show()
```

## Creating two-factor subplots

```
# Create a scatter plot of G1 vs. G3
sns.relplot(x='G1', y='G3',
            data=student_data, kind='scatter')

plt.show()
```

```
# Adjust to add subplots based on school support
sns.relplot(x='G1', y='G3',
            data=student_data, kind='scatter',
            col='schoolsup', col_order=['yes', 'no'])

plt.show()
```

```
# Adjust further to add subplots based on family support
sns.relplot(x='G1', y='G3',
            data=student_data, kind='scatter',
            col='schoolsup', col_order=['yes', 'no'],
            row='famsup', row_order=['yes', 'no'])

plt.show()
```

# **Visualizing a Categorical and a Quantitative Variable**
# Count plots and bar plots
## Count plots

```
survey_data = pd.read_csv(survey_url)
survey_data.head()
```

```
# Create count plot of internet usage
sns.catplot(x='Internet usage', data=survey_data, kind='count', aspect=2);
plt.tight_layout();
```

```
# Create column subplots based on age category
sns.catplot(y="Internet usage", data=survey_data,
            kind='count')
```

```
# Add column 'Age Category' into Dataframe
survey_data.loc[(survey_data.Age<21), 'Age Category'] = 'Less than 21'
survey_data.loc[(survey_data.Age>=21), 'Age Category'] = '21+'

survey_data['Age Category'].head()
```

```
# Create column subplots based on age category
sns.catplot(y="Internet usage", data=survey_data,
            kind="count", col='Age Category')

# Show plot
plt.show()
```

## Bar plots with percentages

```
# Add a column 'Interested in Math' into Dataframe
survey_data.loc[(survey_data.Mathematics>=3), 'Interested in Math'] = True
survey_data.loc[(survey_data.Mathematics<3), 'Interested in Math'] = False
survey_data.loc[(survey_data.Mathematics.isnull()), 'Interested in Math'] = False
```

```
# Create a bar plot of interest in math, separated by gender
sns.catplot(x='Gender', y='Interested in Math',
            data=survey_data, kind='bar')

plt.show()
```

## Customizing bar plots

```
student_data = pd.read_csv(students_url)
student_data.head()
```

```
# Create bar plot of average final grade in each study category
sns.catplot(x='study_time', y='G3',
            data=student_data, kind='bar')

plt.show()
```

```
# Rearrange the categories
sns.catplot(x='study_time', y='G3',
            data=student_data, 
            kind='bar',
            order = [
                     '<2 hours',
                     '2 to 5 hours',
                     '5 to 10 hours', 
                     '>10 hours'
            ])

plt.show()
```

```
# Turn off the confidence intervals
sns.catplot(x='study_time', y='G3',
            data=student_data,
            kind='bar',
            order=[
                   '<2 hours',
                   '2 to 5 hours',
                   '5 to 10 hours',
                   '>10 hours'
            ],
            ci=None)

plt.show()
```

# Box plots
## Create and interpret a box plot

```
# Specify the category ordering
study_time_order = ['<2 hours', '2 tp 5 hours', '5 to 10 hours', '>10 hours']

# Create a box plot and set the order of the categories
sns.catplot(x='study_time', y='G3',
            kind='box', data=student_data,
            order=study_time_order)

plt.show()
```

## Omitting outliers

```
# Create a box plot with subgroups subgroups and omit the outliers
sns.catplot(x='internet', y='G3',
            data=student_data, kind='box',
            hue='location', sym='')

plt.show()
```

## Adjusting the whiskers

```
# Set the whiskers to 0.5 * IOR
sns.catplot(x='romantic', y='G3',
            data=student_data,
            kind='box', whis=0.5)

plt.show()
```

```
# Extend the whiskers to the 5th and 95th percentile
sns.catplot(x='romantic', y='G3',
            data=student_data,
            kind='box', whis=[5, 95])

plt.show()
```

```
# Set the whiskers at the min and max values
sns.catplot(x='romantic', y='G3',
            data=student_data,
            kind='box', whis=[0,100])

plt.show()
```

# Point plots
## Customizing point plots

```
# Create a point plot of family relationship vs. absences
sns.catplot(x='famrel', y='absences',
            data=student_data, kind='point')

plt.show()
```

```
# Add caps to the confidence interval
sns.catplot(x='famrel', y='absences',
            data=student_data,
            kind='point', capsize=0.2)

plt.show()
```

```
# Remove the lines joining the points
sns.catplot(x='famrel', y='absences',
            data=student_data,
            kind='point',
            capsize=0.2, join=False)

plt.show()
```

## Point plots with subgroups

```
# Create apoint plot with subgroups
sns.catplot(x='romantic', y='absences',
            kind='point', data=student_data,
            hue='school'
            )

plt.show()
```

```
# Turn off the confidence intervals for this plot
sns.catplot(x='romantic', y='absences',
            data=student_data,
            kind='point', hue='school', ci=None)

plt.show()
```

```
# Import median function from numpy
from numpy import median

# Plot the median number of absences instead of the mean
sns.catplot(x='romantic', y='absences',
            data=student_data, kind='point',
            hue='school', ci=None, estimator=median)

plt.show()
```

# **Customizing Seaborn Plots**
# Changinig plot style and color
## Changing style and palette

```
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

plt.rcParams['figure.figsize']=[10, 5]
```

```
survey_data = pd.read_csv(survey_url)
survey_data.head()
```

```
mapping = {1:'Never', 2:'Rarely', 3:'Sometimes', 4:'Often', 5:'Always'}
survey_data['Parents Advice']=survey_data['Parents\' advice'].map(mapping)
```

```
# Set the style to 'whitegrid'
sns.set_style('whitegrid')

# Create a count plot of survey responses
category_order = ['Never', 'Rarely', 'Sometimes', 'Often', 'Always']

sns.catplot(x='Parents Advice', data=survey_data, kind='count', order=category_order)

# Show plot
plt.show()
```

```
# Set the color palette to 'Purples'
sns.set_style('whitegrid')
sns.set_palette('Purples')

category_order = ['Never', 'Rarely', 'Sometimes', 'Often', 'Always']

sns.catplot(x='Parents Advice', data=survey_data, kind='count', order=category_order)

plt.show()
```

```
# Change the color palette to 'RdBu'
sns.set_style('whitegrid')
sns.set_palette('RdBu')

category_order=['Never', 'Rarely', 'Sometimes', 'Often', 'Always']

sns.catplot(x='Parents Advice', data=survey_data, kind='count', order=category_order)

plt.show()
```

# Adding a title and axis labels
## Adding a title and axis labels

```
mpg = pd.read_csv(mpg_url)

mpg_mean = mpg.groupby(['model_year', 'origin']).mean().reset_index()

mpg_mean
```

```
# Create lineplot
g = sns.lineplot(x='model_year', y='mpg', data=mpg_mean, hue='origin')

# Add a title 'Average MPG Over Time'
g.set_title('Average MPG Over Time')

plt.show()
```

```
# Create line plot
g = sns.lineplot(x='model_year', y='mpg',
                 data=mpg_mean, hue='origin')

# Add a title 'Average MPG Over Time'
g.set_title('Average MPG Over Time')

# Add x-axis and y-axis labels
g.set(xlabel='Car Model Year',
      ylabel='Average MPG')

plt.show()
```

## Rotating x-tick labels

```
# Create point plot
sns.catplot(x='origin', y='acceleration', data=mpg, kind='point', join=False, capsize=0.1)

# Rotate x-tick labels
plt.xticks(rotation=90)

plt.show()
```

# Putting it all together
## Box plot with subgroups

```
survey_data = pd.read_csv(survey_url)

survey_data['Interested in Pets'] = survey_data['Pets'].apply(lambda x: 'Yes' if x > 4.0 else 'No')

survey_data.head()
```

```
# Set palette to 'Blues'
sns.set_palette('Blues')

# Adjust to add subgroups based on 'Interested in Pets'
g = sns.catplot(x='Gender', y='Age', data=survey_data, 
                kind='box', hue='Interested in Pets')

# Set title to 'Age of Those Interested in Pets vs. Not'
g.fig.suptitle('Age of Those Interested in Pets vs. Not')

plt.show()
```

## Bar plot with subgroups and subplots

```
survey_data['Likes Techno'] = survey_data['Techno'].apply(lambda x: True if x >= 4.0 else False)
survey_data.head()
```

```
# Set the figure style to 'dark'
plt.style.use('seaborn')
sns.set_style('dark')

# Adjust to add subplots per gender
g = sns.catplot(x='Village - town', y='Likes Techno', data=survey_data, kind='bar', col='Gender')

# Add title and axis labels
g.fig.suptitle('Percentage of Young People Who like Techo', y = 1.02)
g.set(xlabel='Location of Residence',
      ylabel='% Who Like Techno')

plt.show()
```
